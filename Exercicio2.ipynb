```python
import pandas as pd
# trocar pelo caminho do seu drive
caminho = '/content/drive/MyDrive/Aula_DataExploratory/DNSE2017.csv'

sinasc = pd.read_csv(caminho, dtype=str)
```

```python
sinasc.QTDFILVIVO = pd.to_numeric(sinasc.QTDFILVIVO).astype(pd.Int64Dtype())
sinasc.QTDFILMORT = pd.to_numeric(sinasc.QTDFILMORT).astype(pd.Int64Dtype())
sinasc.SEMAGESTAC = pd.to_numeric(sinasc.SEMAGESTAC).astype(pd.Int64Dtype())
sinasc.PESO = pd.to_numeric(sinasc.PESO)
sinasc.IDADEMAE = pd.to_numeric(sinasc.IDADEMAE)
sinasc.IDADEPAI = pd.to_numeric(sinasc.IDADEPAI)
sinasc.APGAR1 = pd.to_numeric(sinasc.APGAR1)
sinasc.APGAR5 = pd.to_numeric(sinasc.APGAR5)
sinasc.CONSPRENAT = pd.to_numeric(sinasc.CONSPRENAT).astype(pd.Int64Dtype())
sinasc.SERIESCMAE = pd.to_numeric(sinasc.SERIESCMAE).astype(pd.Int64Dtype())
sinasc.QTDGESTANT = pd.to_numeric(sinasc.QTDGESTANT).astype(pd.Int64Dtype())
sinasc.QTDPARTCES = pd.to_numeric(sinasc.QTDPARTCES).astype(pd.Int64Dtype())
sinasc.QTDPARTNOR = pd.to_numeric(sinasc.QTDPARTNOR).astype(pd.Int64Dtype())
```

```python
#Exercicio 1
partonormal = sinasc[(sinasc.PARTO == '1')]
media_normal = partonormal.IDADEMAE.mean()
dm_normal = partonormal.IDADEMAE.mad() 
variancia_normal = partonormal.IDADEMAE.var()
quartil = partonormal.IDADEMAE.quantile([0.25, 0.75]).values
intervalo_quartil = quartil[1] - quartil[0]
dp_normal = partonormal.IDADEMAE.std()
cv_1 = dp_normal*100/media_normal

resultado = pd.DataFrame({
    'Parto':['Normal'],
    'média':[media_normal], 
    'dm':[dm_normal],
    'var':[variancia_normal],
    'dp':[dp_normal], 
    'iqr':[intervalo_quartil],
    'cv':[cv_1]} ).set_index('Parto')
resultado
```

```python
#Exercicio 3

#valores = sinasc['CONSULTAS'].unique()


pre_natal_1 = sinasc[(sinasc.CONSULTAS == '1')]
pre_natal_2 = sinasc[(sinasc.CONSULTAS == '2')]
pre_natal_3 = sinasc[(sinasc.CONSULTAS == '3')]
pre_natal_4 = sinasc[(sinasc.CONSULTAS == '4')]
#pre_natal_9 = sinasc[(sinasc.CONSULTAS == '9')]

media_1 = pre_natal_1.PESO.mean()
media_2 = pre_natal_2.PESO.mean()
media_3 = pre_natal_3.PESO.mean()
media_4 = pre_natal_4.PESO.mean()

desviop_1 = pre_natal_1.PESO.std()
desviop_2 = pre_natal_2.PESO.std()
desviop_3 = pre_natal_3.std()
desviop_4 = pre_natal_4.std()

desviom_1 = pre_natal_1.PESO.mad()
desviom_2 = pre_natal_2.PESO.mad()
desviom_3 = pre_natal_3.mad()
desviom_4 = pre_natal_4.mad()

q_1 = pre_natal_1.PESO.quantile([0.25, 0.75]).values
q_2 = pre_natal_2.PESO.quantile([0.25, 0.75]).values
q_3 = pre_natal_3.PESO.quantile([0.25, 0.75]).values
q_4 = pre_natal_4.PESO.quantile([0.25, 0.75]).values

iqr_1 = q_1[1] - q_1[0]
iqr_2 = q_2[1] - q_2[0]
iqr_3 = q_3[1] - q_3[0]
iqr_4 = q_4[1] - q_4[0]

cv_1 = desviop_1*100/media_1
cv_2 = desviop_2*100/media_2
cv_3 = desviop_3*100/media_3
cv_4 = desviop_4*100/media_4

resultado = pd.DataFrame({
    'Consultas':['Nenuma', '1 a 3', '4 a 6', '7 ou mais'],
    'média':[media_1, media_2, media_3, media_4], 
    'dm':[desviom_1, desviom_2, desviom_3, desviom_4],
    'dp':[desviop_1, desviop_2, desviop_3, desviop_4], 
    'iqr':[iqr_1, iqr_2, iqr_3, iqr_4],
    'cv':[cv_1, cv_2, cv_3, cv_4]} ).set_index('Consultas')
resultado
```
